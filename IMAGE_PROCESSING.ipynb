```python
from google.colab import files
import cv2
from google.colab.patches import cv2_imshow
uploaded= files.upload()
filesname=list(uploaded.keys())
img=cv2.imread(filesname[0])
resized=cv2.resize(img,(200,200))
print("orginal image")
cv2_imshow(img)
print("resized image")
cv2_imshow(resized)
resized_filename="resized_"+filesname[0]
cv2.imwrite(resized_filename,resized)
files.download(resized_filename)
```

```python
#colour conversion in image processing
import cv2
from google.colab import files
import matplotlib.pyplot as plt
uploaded= files.upload()
name=list(uploaded.keys())[0]
img=cv2.imread(name)
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,binary=cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.title("orginal image")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis("on")
plt.show()
```

```python
import cv2
from google.colab import files
import matplotlib.pyplot as plt
uploaded= files.upload()
filename=list(uploaded.keys())[0]
img=cv2.imread(filename)
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
blurred=cv2.GaussianBlur(gray,(5,5),0)
edges=cv2.Canny(blurred,50,150)
output_filename="edge_segmented.png"
cv2.imwrite(output_filename,edges)
print(f"Edge-segmented image saved as {output_filename}")
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("orginal image")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.subplot(1,2,2)
plt.title("edge segmented image")
plt.imshow(edges,cmap="gray")
plt.axis("off")
plt.show()
files.download(output_filename)
```

```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt
uploaded= files.upload()
filename=list(uploaded.keys())[0]
img=cv2.imread(filename)
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
seed=(100,100)
threshold=10
segmented=np.zeros_like(gray)
seed_value=gray[seed]
for i in range(gray.shape[0]):
  for j in range(gray.shape[1]):
    if abs(gray[i,j]-seed_value)<=threshold:
      segmented[i,j]=255
```
