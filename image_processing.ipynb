```
import cv2
from matplotlib import pyplot as plt
from google.colab import files

# Upload file
uploaded = files.upload()

# Get filename
filename = list(uploaded.keys())[0]

# Read image
img = cv2.imread(filename, 0)

# Check
if img is None:
    print("Error loading image")
else:
    # Calculate the grayscale histogram of the image
    # Parameters:
    # [img]      - Source image wrapped in a list (required by OpenCV)
    # [0]        - Channel index (0 for grayscale image)
    # None       - No mask applied; compute histogram for the entire image
    # [256]      - Number of bins (for intensity levels 0 to 255)
    # [0, 256]   - Range of intensity values to be measured (0 to 255 inclusive)
    hist = cv2.calcHist([img], [0], None, [256], [0, 256])

    plt.figure()
    plt.plot(hist)
    plt.title('Histogram')
    plt.show()
```

```
# Averaging Filter (Simple Mean Filter) in Google Colab

import cv2
from matplotlib import pyplot as plt
from google.colab import files

#  Upload file
uploaded = files.upload()

#  Get filename
filename = list(uploaded.keys())[0]

# Read image in grayscale
img = cv2.imread(filename, 0)

# Check if image loaded successfully
if img is None:
    print("Error: Image not loaded")
else:
    #  Define kernel size (e.g., 3x3)
    kernel_size = 3

    #  Apply averaging filter (simple mean)
    averaged_img = cv2.blur(img, (kernel_size, kernel_size))

    #  Display original and filtered images side by side
    plt.figure(figsize=(10,5))

    plt.subplot(1,2,1)
    plt.imshow(img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1,2,2)
    plt.imshow(averaged_img, cmap='gray')
    plt.title('Averaging Filtered Image')
    plt.axis('off')

    plt.show()
```

```
# Weighted Averaging Filter in Google Colab

import cv2
import numpy as np  # ✅ Import NumPy
from matplotlib import pyplot as plt
from google.colab import files

# 1️⃣ Upload file
uploaded = files.upload()

# 2️⃣ Get filename
filename = list(uploaded.keys())[0]

# 3️⃣ Read image in grayscale
img = cv2.imread(filename, 0)

# Check if image loaded
if img is None:
    print("Error: Image not loaded")
else:
    # 4️⃣ Define a weighted averaging kernel (3x3 example)
    kernel = np.array([[1, 1, 1],
                       [1, 4, 1],
                       [1, 1, 1]], dtype=np.float32)

    # Normalize kernel (sum should be 1)
    kernel = kernel / kernel.sum()

    # 5️⃣ Apply weighted averaging filter
    weighted_img = cv2.filter2D(img, -1, kernel)

    # 6️⃣ Display original and filtered images side by side
    plt.figure(figsize=(10,5))

    plt.subplot(1,2,1)
    plt.imshow(img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1,2,2)
    plt.imshow(weighted_img, cmap='gray')
    plt.title('Weighted Averaging Filter')
    plt.axis('off')

    plt.show()
```

```
# Order Statistics Filters (Minimum, Maximum, Median) in Google Colab

import cv2
import numpy as np
from matplotlib import pyplot as plt
from scipy import ndimage
from google.colab import files

# 1️⃣ Upload image
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# 2️⃣ Read image in grayscale
img = cv2.imread(filename, 0)

if img is None:
    print("Error: Image not loaded")
else:
    # 3️⃣ Define window size (neighborhood)
    size = 3  # 3x3 window

    # 4️⃣ Apply Minimum filter (0th percentile)
    min_filtered = ndimage.minimum_filter(img, size=size)

    # 5️⃣ Apply Maximum filter (100th percentile)
    max_filtered = ndimage.maximum_filter(img, size=size)

    # 6️⃣ Apply Median filter
    median_filtered = cv2.medianBlur(img, ksize=size)

    # 7️⃣ Display all results side by side
    plt.figure(figsize=(15,5))

    plt.subplot(1,4,1)
    plt.imshow(img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1,4,2)
    plt.imshow(min_filtered, cmap='gray')
    plt.title('Minimum Filter')
    plt.axis('off')

    plt.subplot(1,4,3)
    plt.imshow(max_filtered, cmap='gray')
    plt.title('Maximum Filter')
    plt.axis('off')

    plt.subplot(1,4,4)
    plt.imshow(median_filtered, cmap='gray')
    plt.title('Median Filter')
    plt.axis('off')

    plt.show()
```

```
# Sharpening Spatial Filter in Google Colab

import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab import files

# 1️⃣ Upload image
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# 2️⃣ Read image in grayscale
img = cv2.imread(filename, 0)

if img is None:
    print("Error: Image not loaded")
else:
    # 3️⃣ First-order derivative (simple edge detection kernel)
    kernel_first_order = np.array([[0, -1, 0],
                                   [-1, 4, -1],
                                   [0, -1, 0]], dtype=np.float32)
    first_order_img = cv2.filter2D(img, -1, kernel_first_order)

    # 4️⃣ Second-order derivative (Laplacian kernel)
    kernel_second_order = np.array([[0, 1, 0],
                                    [1, -4, 1],
                                    [0, 1, 0]], dtype=np.float32)
    second_order_img = cv2.filter2D(img, -1, kernel_second_order)

    # 5️⃣ Sharpened image by adding Laplacian (second-order) to original
    sharpened_img = cv2.addWeighted(img, 1.0, second_order_img, 1.0, 0)

    # 6️⃣ Display results
    plt.figure(figsize=(15,5))

    plt.subplot(1,4,1)
    plt.imshow(img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1,4,2)
    plt.imshow(first_order_img, cmap='gray')
    plt.title('First-order Derivative')
    plt.axis('off')

    plt.subplot(1,4,3)
    plt.imshow(second_order_img, cmap='gray')
    plt.title('Second-order Derivative')
    plt.axis('off')

    plt.subplot(1,4,4)
    plt.imshow(sharpened_img, cmap='gray')
    plt.title('Sharpened Image')
    plt.axis('off')

    plt.show()
```

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```
