# **Reusable Scatter Plot**

```python
import matplotlib.pyplot as plt

def reusable_scatter_plot(
    x, y,
    x_label="X-axis",
    y_label="Y-axis",
    title="Scatter Plot",
    color='blue',
    marker='o',
    size=50,
    grid=True,
    alpha=0.7,
    figsize=(10, 6),
    save_path=None
):
    """
    Draws a reusable scatter plot.

    Parameters:
    - x, y: Data for X and Y axes
    - x_label, y_label: Axis labels
    - title: Plot title
    - color: Marker color
    - marker: Marker style
    - size: Marker size
    - grid: Whether to show grid
    - alpha: Transparency level of markers
    - figsize: Figure size as tuple
    - save_path: If specified, saves the plot to this path
    """

    plt.figure(figsize=figsize)
    plt.scatter(x, y, c=color, marker=marker, s=size, alpha=alpha)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    if grid:
        plt.grid(True)
    if save_path:
        plt.savefig(save_path, bbox_inches='tight')
    else:
        plt.show()
```

```python
import numpy as np

# Sample data
x = np.random.rand(100)
y = np.random.rand(100)

reusable_scatter_plot(
    x, y,
    # x_label="Random X",
    y_label="Random Y",
    title="My Reusable Scatter Plot",
    color='green',
    marker='x'
)
```

```python
#Clsutering and Recommendations
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
# from sklearn.metrics.pairwise import cosine_similarity


cols = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo','uri','genre','song_name']
filtered_df = df[cols]
filtered_df

num_cols = [i for i in filtered_df.columns if filtered_df[i].dtype != 'object']
scaler = StandardScaler()

filtered_df[num_cols] = scaler.fit_transform(filtered_df[num_cols])
filtered_df

# Perform PCA with your initial higher n_components
n_components = len(filtered_df[num_cols].columns)  # Set the number of components to the number of features

# Fit PCA on the filtered data
pca = PCA(n_components=n_components)
pca_result = pca.fit_transform(filtered_df[num_cols])

# Explained Variance Ratio
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance_ratio.cumsum()

n_clusters = 7 # Number of clusters

kmeans = KMeans(n_clusters=n_clusters, random_state=42)
filtered_df_pca = filtered_df
filtered_df_pca['cluster'] = kmeans.fit_predict(pca_result)

# Visualize PCA Results
plt.figure(figsize=(10, 8))
sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1], hue=filtered_df_pca['cluster'], palette='viridis', s=50)
plt.title(f'PCA Results with K-Means Clustering with {n_clusters} clusters')
plt.show()
```

# **Vertical and Horizontal Bar Charts**

```python
# Sample data
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]

# Vertical Bar Chart
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color='skyblue')
plt.title("Vertical Bar Chart")
plt.xlabel("Category")
plt.ylabel("Value")
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.show()

# Horizontal Bar Chart
plt.figure(figsize=(6, 4))
plt.barh(categories, values, color='lightcoral')
plt.title("Horizontal Bar Chart")
plt.xlabel("Value")
plt.ylabel("Category")
plt.grid(True, axis='x', linestyle='--', alpha=0.5)
plt.show()
```

```python
# df.isnull().sum()
missing_values = df.isnull().sum()

# Filter columns with missing values only
missing_values = missing_values[missing_values > 0]

# Plot missing values
plt.figure(figsize=(12, 6))
sns.barplot(x=missing_values.index, y=missing_values.values, palette="Reds_r")
plt.xlabel("Columns with Missing Values")
plt.ylabel("Number of Missing Entries")
plt.title("Missing Values per Column in the Spotify Dataset")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot count of each feature (number of non-null values per column)
plt.figure(figsize=(16, 6))
sns.barplot(x=df.columns, y=df.count(), palette="viridis")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Non-Null Entries")
plt.title("Feature Completeness in the Dataset")
plt.show()
```

# **Area Chart**

```python
# Example: Monthly sales
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [200, 240, 300, 280, 350]
plt.fill_between(months, sales, color='skyblue', alpha=0.5)
plt.plot(months, sales, color='blue')  # Optional line for visibility
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
# Simulated multi-series data
days = np.arange(1, 11)
values1 = np.random.randint(20, 50, size=10)
values2 = np.random.randint(10, 30, size=10)

df = pd.DataFrame({'Day': days, 'Value1': values1, 'Value2': values2})
df.set_index('Day', inplace=True)

# Area chartea using stacked area plot
df.plot.area(alpha=0.6, figsize=(7, 4), cmap='Accent')
plt.title("Area Chart")
plt.xlabel("Day")
plt.ylabel("Values")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
