```python
#importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
catagories=["A","B","C","D","E"]
values=np.random.rand(len(catagories))
#creating color colormap
colors=plt.cm.viridis(values)
plt.bar(catagories,values,color=colors)
plt.colorbar(plt.cm.scalarmappable(cmap="viridis"),labels="color scale")
plt.title("bar chart with color scale")
plt.show()
```

```python
catagories=["A","B","C","D","E"]
values=np.random.rand(len(catagories))
colors=plt.cm.plasma(values)
plt.bar(catagories,values,color=colors)
plt.colorbar(plt.cm.scalarmappabl(cmap="plasma"),labels="color scale")
plt.title("bar chart with color scale")
plt.show()
```
