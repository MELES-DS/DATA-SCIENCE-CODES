```python
late_students=[3,4,5,2,3,4,6,2,3,1,4,5,3,4,2,3,5,4,6,3]
n=37
proportions_late=[x/n for x in late_students]
p_bar=np.mean(proportions_late)

std=np.std(proportions_late)

ucl=p_bar+3*std

lcl=p_bar-3*std


```

```python
late_student=[3,4,5,2,3,4,6,2,3,1,4,5,3,4,2,3,5,4,6,3]
n=37
proportional_late=[x/n for x in late_student]
# Calculate p-chart parameters
p_bar = np.mean(proportional_late)  # Average proportion
p_bar
std_p = np.sqrt((p_bar * (1 - p_bar)) / n)  # Standard deviation of p
std_p
ucl_p = p_bar + 3 * std_p
ucl_p
lcl_p = p_bar - 3 * std_p
lcl_p
# Create the p-chart
#plt.title('Control Chart of p-chart')
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(proportional_late) + 1), proportional_late, marker='o', linestyle='-', color='blue', label='Proportion Late')
plt.axhline(p_bar, color='green', linestyle='--', label=f'Average Proportion (p̄) = {p_bar:.3f}')
plt.axhline(ucl_p, color='red', linestyle='--', label=f'UCL = {ucl_p:.3f}')
plt.axhline(lcl_p, color='red', linestyle='--', label=f'LCL = {lcl_p:.3f}')
```
