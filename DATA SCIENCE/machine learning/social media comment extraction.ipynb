```python
from googleapiclient.discovery import build
from collections import Counter
import re
import nltk
from nltk.corpus import stopwords

# Ensure you have the necessary NLTK resources
nltk.download('stopwords')

# Your API Key from Google Cloud Console
API_KEY = 'AIzaSyDL5j2VAVWjTVOoSIZKJAdEjuNRF1uvOUk'  # Replace with your actual API key

# List of stopwords
stop_words = set(stopwords.words('english'))

# Function to get YouTube API client
def get_youtube_client():
    youtube = build('youtube', 'v3', developerKey=API_KEY)
    return youtube

# Function to clean comment text (remove emojis, stop-words, and images)
def clean_comment(comment):
    # Remove emojis using regex
    comment = re.sub(r'[^\w\s,]', '', comment)
    
    # Remove stopwords
    words = comment.split()
    cleaned_comment = [word for word in words if word.lower() not in stop_words]
    
    return ' '.join(cleaned_comment)

# Function to get comments from a specific video
def get_video_comments(video_id):
    youtube = get_youtube_client()
    comments = []
    authors = []
    next_page_token = None

    while True:
        # Fetch comment threads (comments associated with a video)
        request = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=100,  # You can adjust this number
            textFormat="plainText",
            pageToken=next_page_token
        )
        response = request.execute()

        # Extract comments and authors
        for item in response["items"]:
            comment = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
            author = item["snippet"]["topLevelComment"]["snippet"]["authorDisplayName"]
            cleaned_comment = clean_comment(comment)  # Clean the comment
            comments.append(cleaned_comment)
            authors.append(author)

        # Pagination: check if there is a next page of comments
        next_page_token = response.get("nextPageToken")
        if not next_page_token:
            break

    return authors, comments

# Function to display the total number of comments and frequency of each author
def display_author_frequency(authors):
    if not authors:
        print("No comments found.")
        return

    # Count frequency of each author
    author_counter = Counter(authors)

    # Display total comments and frequency of each author
    print(f"Total Comments: {len(authors)}\n")
    print("Frequency of Comments by Author:")
    for author, frequency in author_counter.items():
        print(f"Author: {author}, Comments: {frequency}")

# Function to display the cleaned comments
def display_cleaned_comments(comments):
    if not comments:
        print("No comments found.")
        return

    print(f"Cleaned Comments (Total: {len(comments)}):\n")
    for i, comment in enumerate(comments, 1):
        print(f"Comment #{i}: {comment}")
        print("-" * 60)

if __name__ == "__main__":
    VIDEO_ID = "Cyb2u-JcnOM"  # Replace with your actual video ID
    authors, cleaned_comments = get_video_comments(VIDEO_ID)
    
    # Display frequency of authors' comments
    display_author_frequency(authors)
    
    # Display cleaned comments
    display_cleaned_comments(cleaned_comments)
```
