```python
# Clustering Fruits Based on Sweetness, Color, Water Content, Vitamin Content, and Sourness

# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Step 2: Create mock dataset
# Features: sweetness (1-10), color (1-10 for color depth), water_content (%), vitamin_content (1-10), sourness (1-10)
data = {
    'Fruit': ['Apple', 'Banana', 'Orange', 'Mango', 'Grapes', 
              'Strawberry', 'Pineapple', 'Pear', 'Watermelon', 'Blueberry'],
    'Sweetness':       [7, 8, 6, 9, 8, 9, 7, 6, 8, 10],
    'Color':           [5, 6, 7, 8, 7, 9, 6, 5, 4, 10],
    'Water_Content':   [84, 74, 86, 83, 81, 91, 87, 80, 92, 85],
    'Vitamin_Content': [7, 6, 9, 9, 8, 10, 9, 7, 6, 10],
    'Sourness':        [3, 2, 7, 3, 4, 6, 8, 2, 1, 5]
}

df = pd.DataFrame(data)
features = df[['Sweetness', 'Color', 'Water_Content', 'Vitamin_Content', 'Sourness',"clustring"]]

# Step 3: Feature Scaling
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Step 4: KMeans Clustering (let's say we want 3 clusters)
kmeans = KMeans(n_clusters=3, random_state=0)
df['Cluster'] = kmeans.fit_predict(scaled_features)

# Step 5: Visualize in 2D (PCA for dimensionality reduction)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)

plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=df['Cluster'], cmap='tab10', s=100)

for i, name in enumerate(df['Fruit']):
    plt.annotate(name, (pca_result[i, 0], pca_result[i, 1]), fontsize=9)

plt.title("Fruit Clusters (PCA 2D Projection)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.show()

# Step 6: Show final data with clusters
print(df)
```

```python
# Fruit Grouping by Different Properties (Machine Learning)

# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Step 2: Define Dataset (Example Data of 10 Fruits)
# Features: sweetness (1-10), color (scale 1-10), water content (%), vitamin content (1-10), sourness (1-10)
data = {
    'Fruit': ['Apple', 'Banana', 'Orange', 'Mango', 'Grapes', 
              'Strawberry', 'Pineapple', 'Pear', 'Watermelon', 'Blueberry'],
    'Sweetness': [7, 8, 6, 9, 8, 9, 7, 6, 8, 10],
    'Color': [5, 6, 7, 8, 7, 9, 6, 5, 4, 10],
    'Water_Content': [84, 74, 86, 83, 81, 91, 87, 80, 92, 85],
    'Vitamin_Content': [7, 6, 9, 9, 8, 10, 9, 7, 6, 10],
    'Sourness': [3, 2, 7, 3, 4, 6, 8, 2, 1, 5]
}

# Step 3: Create DataFrame
df = pd.DataFrame(data)

# Step 4: Group by Sweetness
grouped_by_sweetness = df.groupby('Sweetness')
print("Grouped by Sweetness:")
for sweetness_level, group in grouped_by_sweetness:
    print(f"\nSweetness Level {sweetness_level}:")
    print(group[['Fruit', 'Sweetness']])

# Step 5: Group by Color
grouped_by_color = df.groupby('Color')
print("\nGrouped by Color:")
for color_value, group in grouped_by_color:
    print(f"\nColor {color_value}:")
    print(group[['Fruit', 'Color']])

# Step 6: Group by Water Content
grouped_by_water = df.groupby('Water_Content')
print("\nGrouped by Water Content:")
for water, group in grouped_by_water:
    print(f"\nWater Content {water}%:")
    print(group[['Fruit', 'Water_Content']])

# Step 7: Group by Vitamin Content
grouped_by_vitamins = df.groupby('Vitamin_Content')
print("\nGrouped by Vitamin Content:")
for vitamin, group in grouped_by_vitamins:
    print(f"\nVitamin Content {vitamin}:")
    print(group[['Fruit', 'Vitamin_Content']])

# Step 8: Group by Sourness
grouped_by_sourness = df.groupby('Sourness')
print("\nGrouped by Sourness:")
for sourness, group in grouped_by_sourness:
    print(f"\nSourness Level {sourness}:")
    print(group[['Fruit', 'Sourness']])

# Step 9: Optional - Bin Sweetness for clearer grouping
df['Sweetness_Level'] = pd.cut(df['Sweetness'], bins=[0,5,7,10], labels=['Low', 'Medium', 'High'])
print("\nGrouped by Binned Sweetness Level:")
print(df.groupby('Sweetness_Level')[['Fruit']].agg(list))

# Step 10: Optional - Feature Scaling (if you want to normalize the features before clustering)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['Sweetness', 'Color', 'Water_Content', 'Vitamin_Content', 'Sourness']])

# Step 11: Plot the fruits based on their Sweetness and Color
plt.figure(figsize=(8, 6))
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=df['Sweetness_Level'].cat.codes, cmap='viridis')
plt.title('Fruit Clustering by Sweetness and Color')
plt.xlabel('Standardized Sweetness')
plt.ylabel('Standardized Color')
plt.grid(True)
plt.show()

# Step 12: View the final DataFrame with Sweetness Level Binned
print("\nFinal DataFrame with Sweetness Level:")
print(df)
```

```python
# Fruit Clustering using KMeans (Machine Learning)

# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Step 2: Define Dataset (Example Data of 10 Fruits)
# Features: weight (grams), size (cm), sweetness (1-10)
data = {
    'Fruit': ['Apple', 'Banana', 'Orange', 'Mango', 'Grapes', 'Strawberry', 'Pineapple', 'Pear', 'Watermelon', 'Blueberry'],
    'Weight': [150, 120, 130, 200, 5, 15, 1000, 160, 3000, 2],
    'Size': [8, 20, 10, 12, 1, 3, 25, 9, 30, 1],
    'Sweetness': [7, 8, 6, 9, 8, 9, 7, 6, 8, 10]
}

df = pd.DataFrame(data)
features = df[['Weight', 'Size', 'Sweetness']]

# Step 3: Feature Scaling (important for KMeans)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Step 4: Apply KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(scaled_features)
df['Cluster'] = clusters

# Step 5: Visualize Clusters (2D using first two features)
plt.figure(figsize=(8,6))
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=clusters, cmap='viridis')
for i, fruit in enumerate(df['Fruit']):
    plt.annotate(fruit, (scaled_features[i, 0], scaled_features[i, 1]))
plt.title('Fruit Clustering by Weight and Size')
plt.xlabel('Standardized Weight')
plt.ylabel('Standardized Size')
plt.grid(True)
plt.show()

# Step 6: View Clustered Data
print(df)
```

```python
# Fruit Clustering with KMeans and Centroid Plotting

# Step 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Step 2: Define the dataset
data = {
    'Fruit': ['Apple', 'Banana', 'Orange', 'Mango', 'Grapes', 
              'Strawberry', 'Pineapple', 'Pear', 'Watermelon', 'Blueberry'],
    'Color': [5, 6, 7, 8, 7, 9, 6, 5, 4, 10],  # scale 110
    'Sweetness': [7, 8, 6, 9, 8, 9, 7, 6, 8, 10],  # scale 110
    'Sourness': [3, 2, 7, 3, 4, 6, 8, 2, 1, 5],   # scale 110
    'Acidity': [4.0, 4.5, 3.2, 3.9, 3.7, 3.0, 3.1, 4.6, 4.8, 3.3],  # pH scale
    'Water_Content': [84, 74, 86, 83, 81, 91, 87, 80, 92, 85],  # percentage
    'Vitamin_Content': [7, 6, 9, 9, 8, 10, 9, 7, 6, 10],  # scale 110
    'Protein_Content': [0.3, 1.1, 0.9, 0.8, 0.6, 0.7, 0.5, 0.4, 0.9, 0.7]  # grams per 100g
}

# Step 3: Create DataFrame
df = pd.DataFrame(data)

# Step 4: Select features and scale them
features = df[['Color', 'Sweetness', 'Sourness', 'Acidity', 'Water_Content', 'Vitamin_Content', 'Protein_Content']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Step 5: Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(scaled_features)
df['Cluster'] = clusters

# Step 6: Add cluster centroids
centroids = kmeans.cluster_centers_

# Step 7: Plot clusters using first two principal features for visualization
plt.figure(figsize=(10, 7))
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=clusters, cmap='viridis', s=100, label='Fruits')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X', label='Centroids')
for i, fruit in enumerate(df['Fruit']):
    plt.annotate(fruit, (scaled_features[i, 0]+0.02, scaled_features[i, 1]+0.02))

plt.title('Fruit Clustering by Color and Sweetness (Scaled)')
plt.xlabel('Standardized Color')
plt.ylabel('Standardized Sweetness')
plt.legend()
plt.grid(True)
plt.show()

# Step 8: Display clustered DataFrame
print("Clustered Fruit Data:")
print(df)
```
